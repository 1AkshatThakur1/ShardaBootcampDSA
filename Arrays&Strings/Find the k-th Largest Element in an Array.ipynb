#Find the k-th Largest Element in an Array

Write a program to find the k-th largest element in a given array of integers. The array may contain duplicates, and you should find the k-th largest distinct element.

Input Format:
The first integer corresponds to n, the number of elements in the array.
The next n integers correspond to the elements of the array.
The last integer corresponds to k, the position of the largest element to find (1-based index).
Output Format:
Print the k-th largest distinct element in the array.
Constraints:
1 ≤ n ≤ 100
-10^5 ≤ array elements ≤ 10^5
1 ≤ k ≤ number of distinct elements in the array
[Refer sample input and output]
[All text in bold refers to the input]

Sample Input 1:

Enter the number of elements:

6

3 2 1 5 6 4

Enter the position to find largest element in Array:

2

Sample Output 1:

5

```python
def find_kth_largest_distinct(arr, k):
    # Convert the list to a set to remove duplicates
    distinct_elements = list(set(arr))

    # Sort the list in descending order
    distinct_elements.sort(reverse=True)

    # Return the k-th largest element (1-based index)
    return distinct_elements[k - 1]

# Input reading
n = int(input("Enter the number of elements:\n"))
print("Enter the elements:")
arr = list(map(int, input().split()))
k = int(input("Enter the position to find largest element in Array:\n"))

# Find the k-th largest distinct element
result = find_kth_largest_distinct(arr, k)

# Output the result
print(result)
```
