#Find the length of the longest subarray with contiguous elements



In the land of Mysteria, the wise sage Amaratha has discovered a mystical artifact, the Array of Elements. Legend has it that this array holds within it the key to unlocking ancient secrets. However, to reveal its true power, one must find the length of the longest subarray with contiguous elements.

Write a program to embark on this quest and uncover the secrets hidden within the Array of Elements.

Input and Output Format:

Input consists of N+1 integers separated by spaces, where the first integer N represents the number of elements in the array, and the following N integers represent the elements of the array. Assume that the maximum value of N is 20.

Output consists of two lines:

    The first line contains a single integer representing the length of the longest subarray with contiguous elements.
    The second line contains a single integer representing the length of the array.

All text in bold corresponds to input and the rest corresponds to output.

[Refer sample input and output]

[All text in bold refers to the input]

Sample Input 1:

Enter the number of elements in the Array of Elements:
6
Enter the mystical array elements:
100 4 200 1 3 2

Sample Output 1:

Behold! The length of the longest contiguous subarray is 3, a sign of the hidden power within.
The sacred array holds within it 6 elements, waiting to be deciphered.

```python
def longest_contiguous_subarray_length(arr):
    n = len(arr)
    max_length = 0

    # Iterate over all subarrays
    for i in range(n):
        min_val = arr[i]
        max_val = arr[i]
        for j in range(i, n):
            min_val = min(min_val, arr[j])
            max_val = max(max_val, arr[j])
            # Check if the current subarray [i, j] is contiguous
            if max_val - min_val == j - i:
                max_length = max(max_length, j - i + 1)

    return max_length

# Read input
N = int(input("Enter the number of elements in the Array of Elements:\n"))
print("Enter the mystical array elements:")
arr = list(map(int, input().split()))

# Find the length of the longest contiguous subarray
longest_length = longest_contiguous_subarray_length(arr)

# Output results
print("Behold! The length of the longest contiguous subarray is", longest_length, ", a sign of the hidden power within.")
print("The sacred array holds within it", N, "elements, waiting to be deciphered.")
```
