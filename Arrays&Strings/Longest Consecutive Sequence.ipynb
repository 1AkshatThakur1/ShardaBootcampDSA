#Longest Consecutive Sequence

Write a program to find the length of the longest consecutive sequence of numbers in an array .

Input and Output Format:

Input consists of N+1 integers separated by spaces, where the first integer N represents the number of elements in the array, and the following N integers represent the elements of the array. Assume that the maximum value of N is 20.

Output consists of two lines:

    The first line contains a single integer representing the length of the longest consecutive sequence.
    The second line contains a single integer representing the length of the array.

All text in bold corresponds to input and the rest corresponds to output.
[Refer sample input and output]

[All text in bold refers to the input]
Sample Input 1:

Enter the number of elements in the array:
6
Enter the array elements:
100

4

200

1

3

2

Sample Output 1:

The length of the longest consecutive sequence is 4

```python
def longest_consecutive_sequence(arr):
    # Convert the list to a set to remove duplicates and allow O(1) time complexity checks
    num_set = set(arr)
    max_length = 0

    # Iterate through the set
    for num in num_set:
        # Check if it is the start of a sequence
        if num - 1 not in num_set:
            current_num = num
            current_length = 1

            # Count the length of the sequence
            while current_num + 1 in num_set:
                current_num += 1
                current_length += 1

            # Update the max length
            max_length = max(max_length, current_length)

    return max_length

# Read input
N = int(input("Enter the number of elements in the array:\n"))
print("Enter the array elements:")
arr = list(map(int, input().split()))

# Find the length of the longest consecutive sequence
longest_length = longest_consecutive_sequence(arr)

# Output results
print(f"The length of the longest consecutive sequence is {longest_length}")
print(f"The length of the array is {N}")
```
