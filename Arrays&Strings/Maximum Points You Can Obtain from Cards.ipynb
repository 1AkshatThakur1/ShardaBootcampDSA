#Maximum Points You Can Obtain from Cards

In this challenge, you are given an array of integers representing card points. You can pick k cards from either the start or the end of the array to maximize the total points. Your task is to find the maximum points you can obtain.

Input Format :
First line: An integer n representing the number of cards.
Second line: n space-separated integers representing the card points.
Third line: An integer k representing the number of cards you can pick.


Output Format :
A single integer representing the maximum points you can obtain.

Constraints :

1 <= n <= 10^5
1 <= k <= n
1 <= card points <= 10^4
Refer Sample Input and Output.

[All text in bold corresponds to input and the rest corresponds to output]

Sample Input and Output :

Enter the number of cards:
7
Enter the card points:
1 2 3 4 5 6 1
Enter the number of cards to pick:
3

12

```python
def maxScore(card_points, k):
    total_points = sum(card_points)
    n = len(card_points)
    window_sum = sum(card_points[:n-k])
    max_sum = total_points - window_sum

    for i in range(n - k, n):
        window_sum += card_points[i] - card_points[i - (n - k)]
        max_sum = max(max_sum, total_points - window_sum)

    return max_sum

def main():
    n = int(input("Enter the number of cards: "))
    print("Enter the card points:")
    card_points = list(map(int, input().split()))
    k = int(input("Enter the number of cards to pick: "))

    max_points = maxScore(card_points, k)
    print(max_points)

if __name__ == "__main__":
    main()
```
