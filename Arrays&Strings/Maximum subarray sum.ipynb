#Maximum subarray sum


Sumit is good in mathematics. His maths teacher has given a sequence of numbers (An array of elements) which includes positive and negative integers. The teacher asked Sumit to find the maximum sum of subarray from the given sequence of numbers. He can consider one element as subarray, also full array as one subarray.
Help Sumit by writing a program using a divide and conquer strategy.


Sample input and output format:
The first line of input refers number of elements(N)
The second line of inputs refers to array elements separated by space.
The third line is an output refer to maximum subarray sum
Refer sample input and output for formatting specifications

[All texts in bold correspond to input and rest corresponds to output]
Sample input and output 1:
5
-1 -2 3 4 5
Maximum subarray sum is 12


Sample input and output 2:
5
1 2 3 4 5
Maximum subarray sum is 15

```python
def max_subarray_sum(arr):
    max_sum = float('-inf')
    current_sum = 0

    for num in arr:
        current_sum = max(num, current_sum + num)
        max_sum = max(max_sum, current_sum)

    return max_sum

def main():
    n = int(input("Enter the number of elements: "))
    arr = list(map(int, input().split()))

    max_sum = max_subarray_sum(arr)

    print("Maximum subarray sum is", max_sum)

if __name__ == "__main__":
    main()
```
