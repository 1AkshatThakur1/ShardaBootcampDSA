#Rotate an array to the right by k steps

In the kingdom of Eldoria, deep within the enchanted forest lies the fabled Treasure Vault, rumored to hold immeasurable riches and ancient relics. However, accessing its treasures isn't easy; one must solve the mystery of rotating the key mechanism to unlock its secrets.

Write a program to assist adventurers in this quest by rotating the key mechanism of the Treasure Vault.

Input and Output Format:

Input consists of two lines:

    The first line contains two integers, N and k, separated by a space, where N represents the number of key pins in the mechanism and k represents the number of steps to rotate the key to the right.
    The second line contains N integers separated by spaces, representing the initial positions of the key pins in the mechanism.

Output consists of a single line:

    The line contains N integers separated by spaces, representing the positions of the key pins after rotating the key to the right by the specified number of steps.

All text in bold corresponds to input and output, and the rest corresponds to instructions and explanations.
[Refer sample input and output]

[All text in bold refers to the input]

Example:

Input:

Enter the number of key pins and the number of steps to rotate:
6 3
4 6 1 3 2 5

Output:

The key positions after rotating to the right by 3 steps: 3 2 5 4 6 1

Constraints:

    1 ≤ N ≤ 100
    1 ≤ k ≤ N

```python
def rotate_right(arr, k):
    n = len(arr)
    k = k % n  # In case k is greater than n
    return arr[-k:] + arr[:-k]

# Input reading
n, k = map(int, input("Enter the number of key pins and the number of steps to rotate:\n").split())
arr = list(map(int, input().split()))

# Rotating the array
rotated_arr = rotate_right(arr, k)

# Output the result
print("The key positions after rotating to the right by", k, "steps:", ' '.join(map(str, rotated_arr)))
```
