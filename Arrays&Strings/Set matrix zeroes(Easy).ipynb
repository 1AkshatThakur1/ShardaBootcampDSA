#Set matrix zeroes(Easy)

Write a program to modify a given m x n matrix. If an element in the matrix is 0, set its entire row and column to 0. The input matrix will contain integers and is guaranteed to be at least 1x1 in size.

Input Format:
The input starts with two integers, m and n, representing the number of rows and columns in the matrix.
The next m * n integers correspond to the elements of the matrix in row-wise order.
Output Format:
Print the modified matrix, row by row, with elements separated by a single space.
[Refer sample input and output]
[All text in bold refers to the input]
Sample Input 1:
Enter the number of row:

3

Enter the number of coloumn:

4

The Matrix is:

1 2 3 4

5 0 6 7

8 9 10 11

Sample Output 1:
The Resultant Matrix is:

1 0 3 4

0 0 0 0

8 0 10 11

```python
def set_matrix_zeroes(matrix, m, n):
    rows_with_zero = set()
    cols_with_zero = set()

    # Step 1: Identify all rows and columns that contain zeros
    for i in range(m):
        for j in range(n):
            if matrix[i][j] == 0:
                rows_with_zero.add(i)
                cols_with_zero.add(j)

    # Step 2: Set the rows to zero
    for i in rows_with_zero:
        for j in range(n):
            matrix[i][j] = 0

    # Step 3: Set the columns to zero
    for j in cols_with_zero:
        for i in range(m):
            matrix[i][j] = 0

    return matrix

# Input reading
m = int(input("Enter the number of rows:\n"))
n = int(input("Enter the number of columns:\n"))

print("The Matrix is:")

matrix = []
for _ in range(m):
    row = list(map(int, input().split()))
    matrix.append(row)

# Modify the matrix
resultant_matrix = set_matrix_zeroes(matrix, m, n)

# Output the result
print("The Resultant Matrix is:")
for row in resultant_matrix:
    print(' '.join(map(str, row)))
```
