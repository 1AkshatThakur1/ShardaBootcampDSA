#Swap Nodes in Pairs

Given an array of integers, swap every two adjacent elements. If the length of the array is odd, leave the last element as it is.


Input Format:

An integer n representing the number of elements in the array.
An array of n integers.
Output Format:

A new array of integers where every two adjacent elements are swapped.
[Refer sample input and output]
[All text in bold refers to the input]
Sample Input 1:
Enter the number of elements in the array:
6
Enter the elements of the array:
1 2 3 4 5 6
Sample Output 1:
2 1 4 3 6 5
Sample Input 2:
Enter the number of elements in the array:
5
Enter the elements of the array:
1 2 3 4 5
Sample Output 2:
2 1 4 3 5

```python
def swap_pairs(arr):
    for i in range(0, len(arr) - 1, 2):
        arr[i], arr[i + 1] = arr[i + 1], arr[i]

def main():
    n = int(input("Enter the number of elements in the array: "))
    print("Enter the elements of the array:")
    arr = list(map(int, input().split()))

    swap_pairs(arr)

    print("The modified array after swapping adjacent elements:")
    print(" ".join(map(str, arr)))

if __name__ == "__main__":
    main()
```
