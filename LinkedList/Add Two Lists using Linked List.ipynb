#Add Two Lists using Linked List

Write a program to add two non-negative integers represented as linked lists. The digits are stored in reverse order, and each node contains a single digit. The function should return the sum as a linked list in the same reverse order format.

Input Format:
The first input line contains an integer n, the number of nodes in the first linked list.
The second input line contains n integers, representing the digits of the first number.
The third input line contains an integer m, the number of nodes in the second linked list.
The fourth input line contains m integers, representing the digits of the second number.
Output Format:
Print the resulting linked list representing the sum, with each digit separated by a space.
[Refer sample input and output]
[All text in bold refers to the input]

Sample Input :

Enter the element for List 1:

3

2 4 3

Enter the element for List 2:

3

5 6 4

Sample Output :

The List is:

7 0 8

```python
class ListNode:
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next

def create_linked_list_from_list(arr):
    head = ListNode(arr[0])
    current = head
    for num in arr[1:]:
        current.next = ListNode(num)
        current = current.next
    return head

def add_two_numbers(l1, l2):
    dummy_head = ListNode(0)
    current = dummy_head
    carry = 0

    while l1 or l2 or carry:
        val1 = l1.val if l1 else 0
        val2 = l2.val if l2 else 0
        total = val1 + val2 + carry
        carry = total // 10
        current.next = ListNode(total % 10)
        current = current.next
        if l1: l1 = l1.next
        if l2: l2 = l2.next

    return dummy_head.next

def print_linked_list(head):
    current = head
    result = []
    while current:
        result.append(current.val)
        current = current.next
    print(' '.join(map(str, result)))

# Input reading
n = int(input("Enter the element for List 1:\n"))
print("Enter the elements for List 1:")
arr1 = list(map(int, input().split()))

m = int(input("Enter the element for List 2:\n"))
print("Enter the elements for List 2:")
arr2 = list(map(int, input().split()))

# Create linked lists
l1 = create_linked_list_from_list(arr1)
l2 = create_linked_list_from_list(arr2)

# Add the two linked lists
result = add_two_numbers(l1, l2)

# Output the result
print("The List is:")
print_linked_list(result)
```
