#Longest Palindromic Substring

Given a string s consisting of only the characters '(' and ')', return the length of the longest balanced substring. A balanced substring is defined as a substring that contains an equal number of '(' and ')' characters, and for every prefix of the substring, the number of '(' characters is greater than or equal to the number of ')' characters.

Input format :
       s : A string consisting of only the characters '(' and ')'. The length of the string is between 1 and 1000 inclusive.

Output format :
An integer representing the length of the longest balanced substring in s.
Constraints :
1 <= len(s) <= 1000
Input and Output Format:
Refer sample input and output for formatting specifications.

[All text in bold corresponds to input and the rest corresponds to output.]

Sample Input and Output :

Enter the string:

  )()())(
Longest balanced substring length: 4

```python
def longest_balanced_substring(s):
    max_length = 0
    stack = [-1]  # Stack to store the indices of '('

    for i, char in enumerate(s):
        if char == '(':
            stack.append(i)
        else:
            # Pop the last '(' index
            stack.pop()
            if not stack:
                stack.append(i)
            else:
                max_length = max(max_length, i - stack[-1])

    return max_length

# Read input from the user
input_string = input("Enter the string:\n")

# Find the length of the longest balanced substring
result = longest_balanced_substring(input_string)

# Print the result
print(f"Longest balanced substring length: {result}")
```
