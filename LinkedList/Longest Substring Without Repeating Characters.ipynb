#Longest Substring Without Repeating Characters

Write a program to find the length of the longest substring without repeating characters using a linked list to manage the characters of the substring.
Constraints:
s consists of English letters, digits, symbols, and spaces.

Your implementation should use the following structure and functions:

struct ListNode to represent a node in the linked list.
createNode function to create a new node with a given character value.
deleteUpTo function to delete nodes from the head of the linked list up to (and including) a specified character value.
existsInList function to check if a character already exists in the current substring (linked list).
Input Format:
A single line containing the string s.
Output Format:
Print the length of the longest substring without repeating characters.

Sample Input and Output 1:
abcabcbb
3
Explanation:
The answer is "abc", with the length of 3.

Sample Input and Output 2:
bbbbb
1
Explanation:
The answer is "b", with the length of 1.

Sample Input and Output 3:
linkedlist
8
Explanation:
The answer is "nkedlist", with the length of 8

```python
class ListNode:
    def __init__(self, char):
        self.char = char
        self.next = None

def createNode(char):
    return ListNode(char)

def deleteUpTo(head, char):
    while head and head.char != char:
        head = head.next
    if head:
        return head.next
    return None

def existsInList(head, char):
    while head:
        if head.char == char:
            return True
        head = head.next
    return False

def lengthOfLongestSubstring(s):
    if not s:
        return 0

    head = None
    tail = None
    max_length = 0
    current_length = 0
    char_set = set()

    for char in s:
        if existsInList(head, char):
            while head and head.char != char:
                char_set.remove(head.char)
                head = head.next
                current_length -= 1
            if head:
                head = head.next
            current_length -= 1

        new_node = createNode(char)
        if tail:
            tail.next = new_node
        tail = new_node
        if not head:
            head = new_node

        char_set.add(char)
        current_length += 1
        if current_length > max_length:
            max_length = current_length

    return max_length

def main():
    s = input("Enter the string: ")
    result = lengthOfLongestSubstring(s)
    print(result)

if __name__ == "__main__":
    main()
```
