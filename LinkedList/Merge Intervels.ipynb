#Merge Intervels
In the realm of Temporia, there exist ancient time gates scattered across the land. Each gate holds the power to unlock hidden knowledge from different epochs. However, to access the full breadth of wisdom, one must merge overlapping time intervals, revealing the synchronized moments of history.

Your task is to write a C program to merge overlapping time intervals and unveil the secrets concealed within the time gates.

Input and Output Format:

Input consists of N+1 lines. The first line contains a single integer N representing the number of time intervals. The following N lines contain two integers separated by a space, representing the start and end times of each interval.

Output consists of two lines:

The first line contains the merged time intervals in ascending order, separated by spaces.
The second line contains a single integer representing the total number of intervals after merging.

All text in bold corresponds to input and the rest corresponds to output.


Refer Sample Input and Output.

[All text in bold corresponds to input and the rest corresponds to output]


Sample Input :

Enter the number of time intervals:
4
Enter the time intervals:
1 3
2 6
5 8
7 10

Sample Output :

The merged time intervals are: 1 8
The total number of intervals after merging is: 1

```python
def merge_intervals(intervals):
    if not intervals:
        return []

    # Sort intervals by their start times
    intervals.sort(key=lambda x: x[0])

    merged = [intervals[0]]
    for current in intervals[1:]:
        last = merged[-1]
        if current[0] <= last[1]:  # Overlapping intervals
            last[1] = max(last[1], current[1])  # Merge
        else:
            merged.append(current)

    return merged

def main():
    n = int(input("Enter the number of time intervals:\n"))
    print("Enter the time intervals:")

    intervals = []
    for _ in range(n):
        start, end = map(int, input().split())
        intervals.append([start, end])

    merged_intervals = merge_intervals(intervals)

    print("The merged time intervals are: ", end="")
    for interval in merged_intervals:
        print(interval[0], interval[1], end=" ")
    print()

    print("The total number of intervals after merging is:", len(merged_intervals))

if __name__ == "__main__":
    main()
```
