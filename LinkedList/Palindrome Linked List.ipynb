#Palindrome Linked List
Write a program to determine whether a linked list is a palindrome or not.

Input Format:

The first line contains integers separated by spaces representing the elements of the linked list.
The second line contains a single integer representing the length of the linked list.
Output Format:

If the linked list is a palindrome, print "Palindrome".
If the linked list is not a palindrome, print "Not Palindrome".
​​​​​​Constraints:

Maximum length of the linked list is 100.
Elements of the linked list are integers.

​[Refer sample input and output]
[All text in bold refers to the input]
Sample Input :
Enter the number of nodes in the linked list:

 5
Enter the characters for each node separated by spaces:

1 2 3 2 1

Sample Output:
Palindrome



```python
class ListNode:
    def __init__(self, value=0, next=None):
        self.value = value
        self.next = next

def create_linked_list(arr):
    if not arr:
        return None
    head = ListNode(arr[0])
    current = head
    for value in arr[1:]:
        current.next = ListNode(value)
        current = current.next
    return head

def reverse_list(head):
    prev = None
    current = head
    while current:
        next_node = current.next
        current.next = prev
        prev = current
        current = next_node
    return prev

def is_palindrome(head):
    if not head or not head.next:
        return True

    # Find the middle of the linked list
    slow, fast = head, head
    while fast and fast.next:
        slow = slow.next
        fast = fast.next.next

    # Reverse the second half
    second_half_start = reverse_list(slow)

    # Compare the first and second half nodes
    first_half, second_half = head, second_half_start
    while second_half:
        if first_half.value != second_half.value:
            return False
        first_half = first_half.next
        second_half = second_half.next

    return True

def main():
    print("Enter the number of nodes in the linked list:")
    n = int(input().strip())
    if n == 0:
        print("Not Palindrome")
        return

    print("Enter the elements for each node separated by spaces:")
    elements = list(map(int, input().strip().split()))

    head = create_linked_list(elements)
    if is_palindrome(head):
        print("Palindrome")
    else:
        print("Not Palindrome")

if __name__ == "__main__":
    main()
```
