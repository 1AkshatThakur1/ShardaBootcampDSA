#Regular Expression Matching

Write a code to implement regular expression matching in the realm of Textlandia. Given an input string s and a pattern p, the task is to determine if the pattern matches the entire input string. The patterns can contain the following special characters:

'.' which matches any single character.
'*' which matches zero or more of the preceding element.
The matching should cover the entire input string (not partial).

Refer sample input and output for formatting specifications.
[All text in bold corresponds to input and the rest corresponds to output.]

Input Format:
A single line containing the input string s.
A single line containing the pattern p.
Output Format:
A single line containing true if the pattern matches the entire input string, otherwise false.
Sample input & output 1:
Enter the input string: aa
Enter the pattern: a
false

Sample input & output 2:
Enter the input string: aa
Enter the pattern: a*
true

```python
def is_match(s, p):
    m, n = len(s), len(p)
    dp = [[False] * (n + 1) for _ in range(m + 1)]

    dp[0][0] = True

    # Handle patterns like a*, a*b*, a*b*c* which can match an empty string
    for j in range(1, n + 1):
        if p[j - 1] == '*':
            dp[0][j] = dp[0][j - 2]

    for i in range(1, m + 1):
        for j in range(1, n + 1):
            if p[j - 1] == '*':
                dp[i][j] = dp[i][j - 2] or (dp[i - 1][j] if (p[j - 2] == s[i - 1] or p[j - 2] == '.') else False)
            else:
                dp[i][j] = dp[i - 1][j - 1] if (p[j - 1] == s[i - 1] or p[j - 1] == '.') else False

    return dp[m][n]

# Read input
input_string = input("Enter the input string: ").strip()
pattern = input("Enter the pattern: ").strip()

# Check if the pattern matches the input string
result = is_match(input_string, pattern)

# Print the result
print("true" if result else "false")
```
