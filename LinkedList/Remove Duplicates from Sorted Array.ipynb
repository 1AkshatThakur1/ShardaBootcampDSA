#Remove Duplicates from Sorted Array

Write a program to remove duplicates from a sorted array of integers and find the length of the new array without duplicates. The program should also print the modified array containing only the unique elements.

Input and Output Format
Input consists of N+1 integers separated by spaces, where the first integer N represents the number of elements in the array, and the following N integers represent the elements of the array. The array is sorted in non-decreasing order. Assume that the maximum value of N is 20.

Output consists of two lines:

The first line contains a single integer representing the length of the array after removing duplicates.
The second line contains the modified array with only unique elements, separated by spaces.

Refer Sample Input and Output.

[All text in bold corresponds to input and the rest corresponds to output]


Sample Input and Output
Sample Input :

Enter the number of elements in the array:
6
Enter the array elements:
1 1 2 2 3 3


Sample Output :

The length of the array after removing duplicates is 3
The modified array is 1 2 3

```python
def remove_duplicates(arr):
    if not arr:
        return 0, []

    unique_arr = [arr[0]]
    for i in range(1, len(arr)):
        if arr[i] != arr[i - 1]:
            unique_arr.append(arr[i])

    return len(unique_arr), unique_arr

def main():
    try:
        n = int(input("Enter the number of elements in the array:\n"))
        if n < 1 or n > 20:
            print("Error: The number of elements must be between 1 and 20.")
            return

        print("Enter the array elements:")
        arr = list(map(int, input().split()))
        if len(arr) != n:
            print("Error: The number of elements does not match the specified size.")
            return

    except ValueError:
        print("Error: Invalid input format.")
        return

    new_length, unique_arr = remove_duplicates(arr)

    print(f"The length of the array after removing duplicates is {new_length}")
    print("The modified array is", " ".join(map(str, unique_arr)))

if __name__ == "__main__":
    main()
```
