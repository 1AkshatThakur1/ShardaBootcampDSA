#Reverse Nodes in k-Group

Given a singly linked list, write a function to reverse nodes in k-group. Given a linked list, reverse the nodes of the list k at a time and return the modified list. If the number of nodes is not a multiple of k, the remaining nodes should be left as is.

Input Format :

The first line contains an integer n, the number of elements in the linked list.
The second line contains n space-separated integers, representing the elements of the linked list.
The third line contains an integer k, the size of the groups to be reversed.
Output Format :

Print the linked list after reversing every k-group of nodes.
[Refer sample input and output]
[All text in bold refers to the input]

Sample Input 1:

Enter the number of elements:

8

Enter the elements:

1 2 3 4 5 6 7 8

Enter the size of the groups to be reversed:

3

Sample Output 1:

Linked list after reversing every group of nodes:

3 2 1 6 5 4 7 8

```python
class ListNode:
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next

def createLinkedList(values):
    if not values:
        return None
    head = ListNode(values[0])
    current = head
    for value in values[1:]:
        current.next = ListNode(value)
        current = current.next
    return head

def printLinkedList(head):
    current = head
    while current:
        print(current.val, end=" " if current.next else "\n")
        current = current.next

def reverseKGroup(head, k):
    if k == 1 or not head:
        return head

    # Check the length of the list
    current = head
    count = 0
    while current:
        current = current.next
        count += 1

    if count < k:
        return head

    # Reverse first k nodes
    prev, current = None, head
    for _ in range(k):
        next_node = current.next
        current.next = prev
        prev = current
        current = next_node

    # Recursively call for the rest of the list
    if current:
        head.next = reverseKGroup(current, k)

    return prev

def main():
    print("Enter the number of elements:")
    n = int(input().strip())

    print("Enter the elements:")
    values = list(map(int, input().strip().split()))

    print("Enter the size of the groups to be reversed:")
    k = int(input().strip())

    head = createLinkedList(values)
    head = reverseKGroup(head, k)

    print("Linked list after reversing every group of nodes:")
    printLinkedList(head)

if __name__ == "__main__":
    main()
```
