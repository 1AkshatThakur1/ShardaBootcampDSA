#Swapping Nodes in a Linked List

You are given a singly linked list and two positions x and y. Write a function to swap the nodes at these positions. The function should update the links in the list to perform the swap. Do not simply swap the data in the nodes.

Input Format:

The first line contains an integer n, the number of elements in the linked list.
The second line contains n space-separated integers, representing the elements of the linked list.
The third line contains two integers x and y, the positions of the nodes to be swapped (0-indexed).
Output Format:

Print the linked list after swapping the nodes at positions x and y.
[Refer sample input and output]
[All text in bold refers to the input]

Sample Input 1:

Enter the number of elements:

6

Enter the elements:

10 15 12 13 20 14

Enter the positions to swap:

2 4

Sample Output 1:

Linked list after swapping nodes:

10 15 20 13 12 14

```python
class ListNode:
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next

def createLinkedList(values):
    if not values:
        return None
    head = ListNode(values[0])
    current = head
    for value in values[1:]:
        current.next = ListNode(value)
        current = current.next
    return head

def printLinkedList(head):
    current = head
    while current:
        print(current.val, end=" " if current.next else "\n")
        current = current.next

def swapNodes(head, x, y):
    if x == y:
        return head

    # Initialize previous and current pointers for x and y
    prevX = prevY = None
    currX = currY = head

    # Find x-th node
    for _ in range(x):
        prevX = currX
        currX = currX.next

    # Find y-th node
    for _ in range(y):
        prevY = currY
        currY = currY.next

    # If x-th or y-th node is not present, return the original list
    if not currX or not currY:
        return head

    # If x is not the head of linked list
    if prevX:
        prevX.next = currY
    else:
        head = currY

    # If y is not the head of linked list
    if prevY:
        prevY.next = currX
    else:
        head = currX

    # Swap next pointers
    currX.next, currY.next = currY.next, currX.next

    return head

def main():
    print("Enter the number of elements:")
    n = int(input().strip())

    print("Enter the elements:")
    values = list(map(int, input().strip().split()))

    print("Enter the positions to swap:")
    x, y = map(int, input().strip().split())

    head = createLinkedList(values)
    head = swapNodes(head, x, y)

    print("Linked list after swapping nodes:")
    printLinkedList(head)

if __name__ == "__main__":
    main()
```
