#Next Greate Element II

Write a code to find the next greater number in a circular array In the land of Arithmos, numbers form circular sequences where the end connects back to the beginning. For each number, the quest is to find the next greater number in the sequence. If no such number exists, the answer should be -1. Let's write a C program to solve this problem.
Input Format:

A single line with space-separated integers representing the circular array nums.
Output Format:

A single line with space-separated integers representing the next greater numbers for each element in the input array.
Refer sample input and output for formatting specifications.

[All text in bold corresponds to input and the rest corresponds to output.]

Sample Input and Output 1:
Enter the numbers in the circular array (space-separated): 1 2 1
2 -1 2
Sample Input and Output 2:
Enter the numbers in the circular array (space-separated): 5 4 3 2 1
-1 5 5 5 5

```python
def nextGreaterElements(nums):
    n = len(nums)
    result = [-1] * n
    stack = []

    for i in range(2 * n):
        while stack and nums[stack[-1]] < nums[i % n]:
            result[stack.pop()] = nums[i % n]
        if i < n:
            stack.append(i)

    return result

def main():
    nums = list(map(int, input("Enter the numbers in the circular array (space-separated): ").split()))
    result = nextGreaterElements(nums)
    print(" ".join(map(str, result)))

if __name__ == "__main__":
    main()
```
