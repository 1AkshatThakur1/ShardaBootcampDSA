#Evaluate Reverse Polish Notation

You are given a string representing an expression in Reverse Polish Notation (RPN). Your task is to evaluate this expression and return the result. The expression is well-formed and only contains integers and the operators +, -, *, and /.

Input Format:

The input consists of a single line containing a space-separated string of tokens representing the RPN expression.
Output Format:

A single line containing the result of evaluating the RPN expression.
Constraints:

The length of the input string is at most 100.
The expression is well-formed.
All tokens are either integers or one of the operators: +, -, *, /.
It is guaranteed that the result of the expression will be an integer.
[Refer sample input and output]
[All text in bold refers to the input]

Sample Input:
2 1 + 3 *

Sample Output:
9

```python
def eval_rpn(tokens):
    stack = []

    for token in tokens:
        if token.isdigit() or (token.startswith('-') and token[1:].isdigit()):
            stack.append(int(token))
        else:
            b = stack.pop()
            a = stack.pop()
            if token == '+':
                stack.append(a + b)
            elif token == '-':
                stack.append(a - b)
            elif token == '*':
                stack.append(a * b)
            elif token == '/':
                # In Python 3, division between integers returns float, so we use //
                stack.append(int(a / b))

    return stack.pop()

# Read input from user
print("Enter the RPN expression:")
expression = input().strip().split()

# Evaluate the RPN expression
result = eval_rpn(expression)

# Print the result
print(f"Result of the RPN expression is: {result}")
```
